```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue ='day')
```

```python
sns.histplot(tips['total_bill'], kde='True', bins = 14)
```

```python
sns.histplot(tips['total_bill'],bins = 14)
```

```python
sns.displot(tips['total_bill'], kde='True', bins = 14)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue = 'sex')
```

```python
sns.stripplot(x='day', y='tip', data=tips, dodge=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', shade=True)
```

```python
# pair plot

sns.pairplot(titanic.select_dtypes(['number']), hue= 'pclass')
```

```python
sns.clustermap(titanic.select_dtypes(['number']).corr(), cmap='coolwarm', annot=True)
```
